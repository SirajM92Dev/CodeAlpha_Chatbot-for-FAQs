```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
```

```python
# FAQs and Answer (Topic: Online payment app)
FAQs = [
    "What payment methods do you accept?",
    "How can I track my order?",
    "What is your return policy?",
    "How do I request a refund?",
    "How long does delivery take?",
    "Can I cancel my order?",
    "How do I reset my account password?",
    "Do you offer international shipping?",
    "How can I contact customer support?",
    "Is cash on delivery available?"
]

Answers = [
    "We accept credit/debit cards, bank transfers, and cash on delivery.",
    "You can track your order from the 'My Orders' section using your order ID.",
    "Products can be returned within 30 days if they are unused and in their original packaging.",
    "Go to your order history, select the order, and click 'Request Refund'.",
    "Standard delivery takes 3–7 business days, depending on your location.",
    "Yes, you can cancel your order before it has been shipped.",
    "Click 'Forgot Password' on the login page and follow the instructions sent to your email.",
    "Yes, we ship to selected countries. Shipping charges vary by destination.",
    "You can contact us via email, live chat, or our customer support phone number.",
    "Yes, cash on delivery is available for eligible locations."
]
```

```python
# print("Original Question: ", user_question)
def clean_question(raw_text):
    text = raw_text.lower()

    tokens = word_tokenize(text)

    stop_words = set(stopwords.words('english'))

    cleaned_text = [
        word for word in tokens
        if word not in stop_words and word not in string.punctuation
    ]

    return " ".join(cleaned_text)
```

```python
cleaned_faqs = [clean_question(faq) for faq in FAQs]
```

```python
user_question = input("Tell us your question: ")
user_question = clean_question(user_question)
# print("Cleaned Question: ", user_question)
```

```python
vectorizer = TfidfVectorizer()

faq_matrix = vectorizer.fit_transform(cleaned_faqs)

user_question_vector = vectorizer.transform([user_question])
```

```python
similarity_scores = cosine_similarity(user_question_vector, faq_matrix)

best_match_index = np.argmax(similarity_scores)

best_score = similarity_scores[0][best_match_index]
```

```python
print(f"{best_score:.2f}")
if(best_score < 0.50):
    print("Dear user, your question is not relevant!")
else:
    print(f"Answer:\t{Answers[best_match_index]}")
```
